```
a = str(input())
if a == a[::-1] :
  print("palindrome")
else :
  print("not palindrome")
```

```
class Solution(object):
    def isPalindrome(self, x):
        return str(x) == str(x)[::-1]

for i in range(int(input())):    
  print(Solution.isPalindrome("", input("Enter Integer or String to check palindrome : ")))
```
